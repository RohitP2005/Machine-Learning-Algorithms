```python
# Importing Modules
from DecisionTree import DecisionTree
import numpy as np
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Importing dataset
data = pd.read_csv("./dataset.csv")
data.head()
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the data
lb_encod = [" education"," self_employed"," loan_status"]
for obj in lb_encod:
    data[obj] = label_encoder.fit_transform(data[obj])

#Drop unecessary columns
data = data.drop(columns=['loan_id'])
data.head()
```

Splitting Dataset

```python
#Splitting Train-Test
x=data.drop(columns=[' loan_status'])
y=data[' loan_status']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1234)
```

```python
# Initialize your DecisionTree
dt = DecisionTree(min_samples_split=2, max_depth=10)

# Fit the model on the training data
dt.fit(xtrain.values, ytrain.values)
```

```python
# Make predictions on the test set
y_pred = dt.predict(xtest.values)
```

```python
#Check Accuracy
np.sum(ytest == y_pred) / len(ytest)
```
